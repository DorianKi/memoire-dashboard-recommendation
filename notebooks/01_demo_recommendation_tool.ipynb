# Interactive Dashboard Recommendation Tool — Demo Notebook

This notebook presents a simplified demonstration of the recommendation logic developed as part of a Master's thesis on interactive dashboard design and strategic decision-making.

The goal is to show how a set of dashboard characteristics can be translated into structured design recommendations.

The full research project is based on an empirical analysis of 35 internal Power BI dashboard pages coded across 29 variables, followed by the extraction of recommendation rules.

## Notebook objectives

This notebook focuses on three goals:

1. Define a simplified recommendation logic inspired by the thesis methodology.
2. Test this logic on several dashboard design scenarios.
3. Display the resulting recommendations in a readable format.

This notebook is intended as a reproducible and portfolio-friendly demonstration rather than the full research environment.

```python
import pandas as pd
from pprint import pprint

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 120)
```

## Simplified recommendation logic

The function below reproduces a simplified version of the reasoning used in the prototype.

It does not aim to reproduce the full empirical model, but it illustrates how dashboard inputs can be transformed into layout and usability recommendations.

```python
import sys
from pathlib import Path

project_root = Path.cwd().resolve().parent
src_path = project_root / "src"

if str(src_path) not in sys.path:
    sys.path.append(str(src_path))

from reco_tool.rules_engine import recommend_dashboard
```

## Demonstration scenarios

The following scenarios simulate common dashboard design situations inspired by the thesis corpus:

- an executive dashboard focused on KPIs and trends
- a monitoring dashboard combining KPIs, trends, and detailed tables
- a dense dashboard with a higher risk of cognitive overload

```python
scenarios = [
    {
        "scenario": "Executive overview dashboard",
        "kpi_count": 4,
        "has_time_series": True,
        "has_table": False,
        "has_alerts": True,
        "filter_count": 2,
        "info_density": "Moderate",
        "color_hierarchy": "Moderate",
        "size_hierarchy": "High"
    },
    {
        "scenario": "Monitoring dashboard with detail table",
        "kpi_count": 3,
        "has_time_series": True,
        "has_table": True,
        "has_alerts": False,
        "filter_count": 4,
        "info_density": "Moderate",
        "color_hierarchy": "Moderate",
        "size_hierarchy": "Moderate"
    },
    {
        "scenario": "Dense dashboard with overload risk",
        "kpi_count": 5,
        "has_time_series": True,
        "has_table": True,
        "has_alerts": True,
        "filter_count": 5,
        "info_density": "High",
        "color_hierarchy": "Low",
        "size_hierarchy": "Low"
    }
]
```

## Run recommendations

```python
results = []

for scenario in scenarios:
    recommendation = recommend_dashboard(
        kpi_count=scenario["kpi_count"],
        has_time_series=scenario["has_time_series"],
        has_table=scenario["has_table"],
        has_alerts=scenario["has_alerts"],
        filter_count=scenario["filter_count"],
        info_density=scenario["info_density"],
        color_hierarchy=scenario["color_hierarchy"],
        size_hierarchy=scenario["size_hierarchy"]
    )
    
    results.append({**scenario, **recommendation})

df_results = pd.DataFrame(results)
df_results
```

## Readable recommendation view

The next cell prints one recommendation block per scenario for easier reading.

```python
for _, row in df_results.iterrows():
    print("=" * 80)
    print(f"SCENARIO: {row['scenario']}")
    print("-" * 80)
    print(f"Recommended layout: {row['recommended_layout']}")
    print(f"KPI placement: {row['kpi_placement']}")
    print(f"Time series placement: {row['time_series_placement']}")
    print(f"Recommended time series visual: {row['recommended_time_series_visual']}")
    print(f"Table placement: {row['table_placement']}")
    print(f"Filter placement: {row['filter_placement']}")
    print(f"Alerts placement: {row['alerts_placement']}")
    print(f"Overload risk: {row['overload_risk']}")
    print(f"Expected affordance level: {row['expected_affordance_level']}")
    print()
```
